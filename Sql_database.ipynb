```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
```

```python
cbb = pd.read_csv("Resources/cbb.csv", encoding="utf-8")
cbb = cbb.rename(columns={'2P_O':'TWOP_O','2P_D':'TWOP_D','3P_O':'THREEP_O','3P_D':'THREEP_D'})
cbb
```

```python
cbb20 = pd.read_csv("Resources/2020Predictions.csv", encoding="utf-8")
cbb20 = cbb20.drop(["Unnamed: 0"], axis=1)
cbb20 = cbb20.rename(columns={'2P_O':'TWOP_O','2P_D':'TWOP_D','3P_O':'THREEP_O','3P_D':'THREEP_D'})
cbb20
```

```python
cbb20_tourney_Predictions = pd.read_csv("Resources/2020TournamentPredictions.csv", encoding="utf-8")
cbb20_tourney_Predictions = cbb20_tourney_Predictions.drop(['Unnamed: 0', 'index'], axis=1)
cbb20_tourney_Predictions = cbb20_tourney_Predictions.rename(columns={'2P_O':'TWOP_O','2P_D':'TWOP_D','3P_O':'THREEP_O','3P_D':'THREEP_D'})
cbb20_tourney_Predictions
```

```python
lat_lng = pd.read_csv("Resources/Cbb_data2.csv", encoding="utf-8")
lat_lng = lat_lng.drop(['Unnamed: 0', 'index', '3P_O.1'], axis=1)
lat_lng = lat_lng.rename(columns={'CONF_x':'CONF','2P_O':'TWOP_O','2P_D':'TWOP_D','3P_O':'THREEP_O','3P_D':'THREEP_D'})
lat_lng
```

```python
cbb20_df = pd.merge(cbb20_tourney_Predictions, lat_lng[['TEAM', 'LONGITUD', 'LATITUDE']], on='TEAM', how='left')

cbb20_df
```

```python
Base = declarative_base()
```

```python
engine = create_engine("sqlite:///output/MarchMadnessML.sqlite")
```

```python
cbb.to_sql(name='All_Teams', con=engine, if_exists="append", index=False)
```

```python
cbb20.to_sql(name='Postseason_Predictions', con=engine, if_exists="append", index=False)
```

```python
cbb20_df.to_sql(name='Tournament_Predictions', con=engine, if_exists="append", index=False)
```
